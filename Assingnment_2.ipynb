```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('/content/dataset2.csv')
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

#UNIVARIATE ANALYSIS
plt.figure(figsize=(8, 6))
sns.histplot(df['living area'], kde=True)
plt.xlabel('Numeric Column')
plt.ylabel('Frequency')
plt.title('Histogram of living area')
plt.show()
```

```python

numeric_summary = df['living area'].describe()
print(numeric_summary)
```

```python

plt.figure(figsize=(8, 6))
sns.boxplot(x='number of bedrooms', y='number of bathrooms', data=df)
plt.xlabel('number of bedrooms')
plt.ylabel('number of bathrooms')
plt.title('Box Plot of Numeric Column by Category')
plt.show()
```

```python
# Create a countplot for a categorical variable
plt.figure(figsize=(8, 6))
sns.countplot(x='number of bedrooms', data=df)
plt.xlabel('number of bedrooms')
plt.ylabel('Count')
plt.title('Countplot of number of bedrooms')
plt.xticks(rotation=45)
plt.show()
```

```python
#BI-VARIATE


plt.figure(figsize=(8, 6))
plt.scatter(df['number of bedrooms'], df['number of bathrooms'])
plt.xlabel('number of bedrooms')
plt.ylabel('number of bathrooms')
plt.title('Scatter Plot between number of bedrooms  and number of bathrooms')
plt.show()
```

```python
import seaborn as sns

# Create a pair plot for multiple numeric variables
sns.pairplot(df[['number of bedrooms','number of bathrooms', 'living area']])
plt.show()
```

```python
# Calculate the correlation matrix
corr_matrix = df.corr()

# Create a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Create a box plot to compare a numeric variable by category
plt.figure(figsize=(8, 6))
sns.boxplot(x='number of bedrooms', y='number of bathrooms', data=df)
plt.xlabel('number of bedrooms')
plt.ylabel('Number of bathrooms')
plt.title('Box Plot of Numeric Variable by Category')
plt.xticks(rotation=45)
plt.show()
```

```python
import seaborn as sns

# Create a multivariate box plot
plt.figure(figsize=(10, 8))
sns.boxplot(data=df, x='number of bedrooms', y='number of bathrooms')
plt.title('Multivariate Box Plot')
plt.xticks(rotation=45)
plt.show()
```

```python
# Count non-missing values for each column
count_non_missing = df.count()
print(count_non_missing)
```

```python
# Calculate mean, median, and mode for a specific numeric column
mean_value = df['number of bedrooms'].mean()
median_value = df['number of bathrooms'].median()
mode_value = df['lot area'].mode().values[0]

print(f"Mean: {mean_value}")
print(f"Median: {median_value}")
print(f"Mode: {mode_value}")
```

```python
# Check for missing values
missing_values = df.isnull().sum()

# Display columns with missing values
print(missing_values[missing_values > 0])
```

```python
# Remove rows with missing values (use inplace=True to modify the DataFrame)
df.dropna(inplace=True)
```

```python
# Impute missing values in a numeric column with the mean
df['number of bedrooms'].fillna(df['number of bedrooms'].mean(), inplace=True)
```

```python
# Impute missing values in a categorical column with the mode
df['waterfront present'].fillna(df['waterfront present'].mode().values[0], inplace=True)
```
